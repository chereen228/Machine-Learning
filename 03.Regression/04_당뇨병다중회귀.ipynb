```python
# 당뇨병 데이터셋 
# LinearRegression과 DecisionTreeRegressor 이용
# y_test, 다중선형회귀, 결정트리회귀 예측값 (test_size=0.1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

- 산점도와 선형회귀 직선

```python
fig, axs = plt.subplots(figsize=(20,10), ncols=5, nrows=2)
features = ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4','s5', 's6']
for i, feature in enumerate(features):
    row, col = i//5, i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- 각 feature 별 target과의 상관관계

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
for feature in diabetes.feature_names:
    X = df[feature].values.reshape(-1,1)
    y = diabetes.target
    lr = LinearRegression()
    lr.fit(X, y)
    r2 = lr.score(X, y)
    print(f'{feature}: {r2:.4f}')
```

- train/test 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    diabetes.data, diabetes.target, 
    test_size=0.1, random_state=2021
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

- LinearRegression 모델

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)
```

```python
# Weight
w = lr.coef_
w
```

```python
# Bias
b = lr.intercept_
b
```

- DecisionTreeRegressor 모델

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state=2021)
dtr.fit(X_train, y_train)
pred_dtr = dtr.predict(X_test)
```

- 결과 비교

```python
# mse: 오차 제곱 평균
from sklearn.metrics import mean_squared_error
mse_lr = mean_squared_error(y_test, pred_lr)
mse_dtr = mean_squared_error(y_test, pred_dtr)
print(f'MSE - LR: {mse_lr:.4f}, DTR: {mse_dtr:.4f}')
```
